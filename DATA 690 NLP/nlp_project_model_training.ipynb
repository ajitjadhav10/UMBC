```python
df = pd.read_csv("/content/gdrive/MyDrive/UMBC/DATA 690 NLP/Data files/test_2_edited.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={"text":"target_text", "ctext":"source_text"})
df = df[['source_text', 'target_text']]
```

```python
df.head()
```

```python
df['source_text'] = "summarize: " + df['source_text']
df
```

```python
train_df, test_df = train_test_split(df, test_size=0.3,random_state=42)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_df[:5000],
            eval_df=test_df[:100], 
            source_max_token_len=512, 
            target_max_token_len=150, 
            batch_size=8, max_epochs=2, use_gpu=True)
```
